```python
nodes = graph.number_of_nodes();
edges = graph.number_of_edges();
```

```python
degree_dict ={};
```

```python
# Calculate degree for graphs and print it
for i in range(1,nodes):
    degree_dict[i] = graph.degree(i);
print(degree_dict);
```

```python
import operator
sorted_degree = sorted(degree_dict.items(), key=lambda x: x[0])
print(sorted_degree);
```

```python
mat = nx.adjacency_matrix(graph)
```

```python
graph_mat = (mat.todense())
```

```python
##Draw Degree Graph
plt.loglog(sorted_degree, 'b-', marker='o')
plt.title("Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")

plt.axes([0.45, 0.45, 0.45, 0.45])
Gcc = sorted(nx.connected_component_subgraphs(graph), key=len, reverse=True)[0]
pos = nx.spring_layout(Gcc)
plt.axis('off')
nx.draw_networkx_nodes(Gcc, pos, node_size=20)
nx.draw_networkx_edges(Gcc, pos, alpha=0.4)

plt.show()
```

```python
pos = nx.spring_layout(graph, iterations=200)
nx.draw(graph, pos, node_color=range(nodes), node_size=500, cmap=plt.cm.Blues)
plt.show()
```

```python
print("Node Degree")
for v in graph:
    print('%s %s' % (v, graph.degree(v)))

nx.draw_circular(graph, with_labels=True)
plt.show()
```

```python
print("Betweenness")
b = nx.betweenness_centrality(graph)
for v in graph.nodes():
    print("%0.2d %5.3f" % (v, b[v]))

print("Degree centrality")
d = nx.degree_centrality(graph)
for v in graph.nodes():
    print("%0.2d %5.3f" % (v, d[v]))

print("Closeness centrality")
c = nx.closeness_centrality(graph)
for v in graph.nodes():
    print("%0.2d %5.3f" % (v, c[v]))

nx.draw(graph)
plt.show()
```

```python
#Shortest Path
pathlengths = []

print("source vertex {target:length, }")
for v in graph.nodes():
    spl = dict(nx.single_source_shortest_path_length(graph, v))
    print('{} {} '.format(v, spl))
    for p in spl:
        pathlengths.append(spl[p])

print('')
print("average shortest path length %s" % (sum(pathlengths) / len(pathlengths)))

# histogram of path lengths
dist = {}
for p in pathlengths:
    if p in dist:
        dist[p] += 1
    else:
        dist[p] = 1

print('')
print("length #paths")
verts = dist.keys()
for d in sorted(verts):
    print('%s %d' % (d, dist[d]))

print("radius: %d" % nx.radius(graph))
print("diameter: %d" % nx.diameter(graph))
print("eccentricity: %s" % nx.eccentricity(graph))
print("center: %s" % nx.center(graph))
print("periphery: %s" % nx.periphery(graph))
print("density: %s" % nx.density(graph))

nx.draw(graph, with_labels=True)
plt.show()
```
