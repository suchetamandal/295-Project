```python
def getEdgePP(G,node1, node2, pp_dict):
    if (G.has_edge(node1, node2)):
        pp1 = pp_dict.get(node1);
        pp2 = pp_dict.get(node2);
        if pp1 is None :
            pp1= 0;
        if pp2 is None :
            pp2 = 0;
        pp = '%.2f' % round(((pp1 + pp2)/20),2);
        return pp;
    else:
        return 0.0;
```

```python
## Add weight to all edges of the Graph woth average propagation probability
def addPPToEdges(G,pp_dict):
    for e in G.edges():
        G[e[0]][e[1]]['weight'] = getEdgePP(G,e[0],e[1],pp_dict);
    return G
```

```python
def getBetweenness(graph):
    b = nx.betweenness_centrality(graph)
    between_dict={}
    for v in graph.nodes():
        between_dict[v] = b[v]
    return between_dict;
```

```python
def getLeaders(graph,n):
    between_dict = getBetweenness(graph);
    sorted_between = sorted(between_dict.items(), key=lambda x: x[0])
    return list(sorted_between)[:n]
```

```python
def shortest_path(G, source):
    ## Find paths from all nodes co-accessible to the target.
    with nx.utils.reversed(G):
        paths=single_source_maximal_path(G,source,weight = 'weight')       
    return paths
```

```python
import heapq
from heapq import heappush, heappop
from itertools import count
def single_source_maximal_path(G,source,target=None,cutoff=None,weight='weight'):
    dist={}
    paths = {source[0]:[source[0]]}
    seen = {source[0]:0}
    fringe=[]
    heapq.heappush(fringe,(0,source))
    while fringe:
        (d,v)=heapq.heappop(fringe)
        if isinstance(v, (int, np.integer)):
            vert = v
        else:
            vert = round(v[0],4);
        if vert in dist:
            continue # already searched this node.   
        dist[vert] = d
        if vert == target:
            break    
        for w,edgedata in G[vert].items():
            vw_dist = float(dist[vert]) + float(edgedata.get(weight,1))
            if cutoff is not None:
                if vw_dist>cutoff:
                    continue
            if w in dist:
                if vw_dist < dist[w]:
                    raise ValueError('Contradictory paths found:',
                                     'negative weights?')
            elif w not in seen or vw_dist > seen[w]:
                seen[w] = vw_dist
                heapq.heappush(fringe,(vw_dist,w))
                paths[w] = paths[vert]+[w]
    return (dist,paths)
```
